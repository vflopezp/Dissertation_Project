```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score,mean_absolute_error
import matplotlib.pyplot as plt
import wandb
import scikitplot as skplt
from yellowbrick.regressor import PredictionError, ResidualsPlot
from sklearn.preprocessing import StandardScaler
import cpi
import seaborn as sb
cpi.update()
import warnings
warnings.filterwarnings("ignore")
```

```python
# load data
data = pd.read_excel('./final_cleaned_dataset.xlsx')
data = pd.get_dummies(data, drop_first=True)
```

```python
# Use CPI index to adjust cost of engineering services for inflation
data['Year of the project'] = pd.to_datetime(data['Year of the project'], format='%Y') 
data["Cost of engineering services"] = data.apply(
    lambda x: cpi.inflate(x["Cost of engineering services"], x["Year of the project"]), axis=1
)
data['Year of the project'] = data['Year of the project'].dt.year.astype('int32')
data = data.drop(['Year of the project'], axis=1)
```

```python
# split data in test and train
X = data.drop('Cost of engineering services', axis=1) # Features
y = data['Cost of engineering services'] 


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# apply tranformations
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
y_train = sc.fit_transform(y_train.to_frame())
y_test = sc.transform(y_test.to_frame())
```

```python
# create base models
rf = RandomForestRegressor()
mlr = LinearRegression()
svr = SVR()
```

```python
# optimize and fit random forest 

# tested hyperameters
rf_params = {
    'n_estimators': [50, 100, 150, 200, 250, 300, 350, 450, 550, 650, 700],
    'max_features' : [ 1, 3, 5, 7, 9, 11,12, 13, 14, 15, 16]
}

# optimize and fit the model
rf_optimized = GridSearchCV(estimator = rf, param_grid = rf_params, cv = 5, verbose=3,  n_jobs = -1, scoring="r2")

rf_optimized.fit(X_train, y_train)
```

```python
# optimize and fit svr 
svr_params = {
    'C': [ 100, 110, 120, 150, 200, 300, 400, 500, 600 ],
    'epsilon' : [100, 9, 6, 5, 3, 2, 1, 0.5, 0.2, 0.1, 0.14, 0.12, 0.1, 0.08, 0.06, 0.04, 0.02, 0.01, 0.005, 0.001]
}

svr_optimized = GridSearchCV(estimator = svr, param_grid = svr_params, cv = 5, verbose=3,  n_jobs = -1, scoring="r2")
svr_optimized.fit(X_train, y_train)
```

```python
# fiting MLR 
mlr.fit(X_train, y_train)
```

```python
# evaluation function 
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    test_labels = sc.inverse_transform(test_labels.reshape(-1, 1))
    predictions = sc.inverse_transform(predictions.reshape(-1, 1))

    r2 = r2_score(test_labels, predictions)
    errors = abs(predictions - test_labels)
    print('Model Performance')
    print(f'RMSE: {mean_squared_error(test_labels, predictions, squared = False)}.')
    print(f'MAPE: {mean_absolute_percentage_error(test_labels, predictions)}%.')
    print(f'MAE: {mean_absolute_error(test_labels, predictions)}.')
    print(f'R2 = {r2}. \n')

    
```

```python
# Evaluating the performance of the best models on test set
best_rf = rf_optimized.best_estimator_
evaluate(best_rf, X_test, y_test)

best_svr = svr_optimized.best_estimator_
evaluate(best_svr, X_test, y_test)

evaluate(mlr, X_test, y_test)
```
